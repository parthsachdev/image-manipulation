```python
from PIL import Image
import numpy as np

image = Image.open('opera-house.jpg')
image.show()
data = np.asarray(image)

print(data.shape)

# array to Image object
image2 = Image.fromarray(data)
image2.format, image2.mode, image2.size
image2.save('opera-house.png', format='PNG')
```

<code> Convert to Grayscale </code>

```python
gs_image = image.convert(mode='L')
gs_image.show()
gs_image.mode, gs_image.size
```

<code> Resize image </code>

```python
print(image.size)
image.thumbnail((100, 100))
print(image.size)  # aspect ratio is maintained
```

<code> Resize image and ignore the aspect ratio </code>

```python
image = Image.open('opera-house.jpg')
image = image.resize((200, 200))
print(image.size)
```

<code> Flip, Rotate and Crop Images </code>

```python
image = Image.open('opera-house.jpg')
hoz_flip = image.transpose(Image.FLIP_LEFT_RIGHT)
ver_flip = image.transpose(Image.FLIP_TOP_BOTTOM)

pyplot.subplot(311)
pyplot.imshow(image)
pyplot.subplot(312)
pyplot.imshow(hoz_flip)
pyplot.subplot(313)
pyplot.imshow(ver_flip)

pyplot.show()

# Note that imshow can plot images directly without converting it to numpy array
```

```python
image = Image.open('opera-house.jpg')

pyplot.subplot(311)
pyplot.imshow(image)
pyplot.subplot(312)
pyplot.imshow(image.rotate(45))
pyplot.subplot(313)
pyplot.imshow(image.rotate(90))

pyplot.show()
```

```python
image = Image.open('opera-house.jpg')
cropped_image = image.crop((100, 100, 200, 200))   # top left and bottom right corner
cropped_image.show()
cropped_image.size
```
